```python
from prophet import Prophet
import yfinance as yf
```

```python
df =yf.download('BTC-USD',"2018-01-01","2024-08-07")
```

```python
df.head()
```

```python
df = df[['Close']]
df = df.reset_index()
df.head()
```

```python
df.columns = ['ds','y']
df.head()
```

```python
model = Prophet()
model.fit(df)
```

```python
gelecek = model.make_future_dataframe(periods=365)
tahmin = model.predict(gelecek)
```

```python
model.plot(tahmin)
```

```python
model.plot_components(tahmin)
```
