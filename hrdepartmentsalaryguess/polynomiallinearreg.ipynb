```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
df = pd.read_csv("polynomial.csv",sep=';')
pol_reg=PolynomialFeatures(degree=2) #linear regressionun derecesini belirtiyor 2 
```

```python
plt.scatter(df['deneyim'], df['maas'])
plt.xlabel('deneyim')
plt.ylabel('maaş')
plt.show()
#bu grafiğe bakarak bu grafiğin düz olmadığını görüyoruz
```

```python
X = df[['deneyim']]  # Features (experience)
y = df['maas']       # Target variable (salary)

# Create a linear regression model
reg = LinearRegression()

# Fit the model to the data
reg.fit(X, y)

# Predict the values
predicted_values = reg.predict(X)

# Plot the scatter plot
plt.scatter(X, y, label='Data Points')

# Plot the regression line
plt.plot(X, predicted_values, color='red', label='Linear Regression')

# Add labels and legend
plt.xlabel('Experience (deneyim)')
plt.ylabel('Salary (maas)')
plt.title('Linear Regression: Experience vs Salary')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
polynomial_reg=PolynomialFeatures(degree=8) # verdim dezavantajı ise 
x_polynomial=polynomial_reg.fit_transform(df[['deneyim']])
reg=LinearRegression()
reg.fit(x_polynomial,df['maas'])
```

```python
y_head=reg.predict(x_polynomial)
plt.plot(df['deneyim'], y_head, color="red", label="polynomial_reg")
plt.legend()
plt.scatter(df['deneyim'],df['maas'])
plt.show()
```
