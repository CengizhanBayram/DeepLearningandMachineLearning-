```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
df = pd.read_csv("polynomial.csv",sep=';')
pol_reg=PolynomialFeatures(degree=2) #linear regressionun derecesini belirtiyor 2 
```

```python
plt.scatter(df['deneyim'], df['maas'])
plt.xlabel('deneyim')
plt.ylabel('maaş')
plt.show()
#bu grafiğe bakarak bu grafiğin düz olmadığını görüyoruz
```

```python
reg =LinearRegression
reg.fit(df[['deneyim']], df['maas'])
plt.xlabel('deneyim')
plt.ylabel('maaş')
plt.scatter(df['deneyim'], df['maas'])
x=df['deneyim']
y=reg.predict(df['deneyim'])
plt.plot(x,y,color="red", label="linear reg")
plt.legend()
plt.show()
```
