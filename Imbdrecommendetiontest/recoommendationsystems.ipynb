```python
import numpy as np 
import pandas as pd 
```

```python
column_names = ['item_id', 'user_id', 'rating', 'timestamp']
df = pd.read_csv('users.data', sep='\t', names=column_names )
df.head()
```

```python
movie_titles = pd.read_csv("movie_id_titles.csv")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on = ('item_id'))
df.head()
#bir çeşit join yaptı
```

```python
moviemat = df.pivot_table(index='user_id',columns='title', values='rating')
moviemat.head()
```

```python
type(moviemat)
```

```python
film_user_ratings = moviemat['Absolute Power (1997)']
film_user_ratings.head()
```

```python
similar_to_film = moviemat.corrwith(film_user_ratings)
similar_to_film
```

```python
type(similar_to_film)
```

```python
corr_film = pd.DataFrame(similar_to_film, columns=['correlaitons'])
corr_film.dropna(inplace=True)
corr_film.head()
```

```python
corr_film.sort_values('correlaitons', ascending=False).head(10) 
#alakasız çıkması yüzünden oy oranına göre sıralamamız lazım
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.sort_values('rating',ascending=False).head()
```

```python
ratings ['rating_oy_sayısı'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.sort_values('rating',ascending=False).head()
```

```python
ratings.sort_values('rating_oy_sayısı' , ascending=False).head()
```

```python
corr_film=corr_film.join(ratings['rating_oy_sayısı'])
corr_film.head()
```
