```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv(r'c:\Users\cengh\Downloads\diabetes.csv')
data.head()
```

```python

seker_hastalari = data[data.Outcome == 1]
saglikli_insanlar = data[data.Outcome == 0]


# imdilik sadece gloucose'a bakarak rnek bir izim yapalm:
# Programmzn sonunda makine renme modelimiz sadece glikoza deil tm dier verilere bakarak bir tahmin yapacaktr..
plt.scatter(saglikli_insanlar.Age, saglikli_insanlar.Glucose, color="green", label="salkl", alpha = 0.4)
plt.scatter(seker_hastalari.Age, seker_hastalari.Glucose, color="red", label="diabet hastas", alpha = 0.4)
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.legend()
plt.show()


```

```python


# x ve y eksenlerini belirleyelim
y = data.Outcome.values
x_ham_veri = data.drop(["Outcome"],axis=1)   
# Outcome stununu(dependent variable) karp sadece independent variables brakyoruz
# kn KNN algoritmas x deerleri ierisinde gruplandrma yapacak..


# normalization yapyoruz - x_ham_veri ierisindeki deerleri sadece 0 ve 1 arasnda olacak ekilde hepsini gncelliyoruz
# Eer bu ekilde normalization yapmazsak yksek rakamlar kk rakamlar ezer ve KNN algoritmasn yanltabilir!
x = (x_ham_veri - np.min(x_ham_veri))/(np.max(x_ham_veri)-np.min(x_ham_veri))

# nce
print("Normalization ncesi ham veriler:\n")
print(x_ham_veri.head())


# sonra 
print("\n\n\nNormalization sonras yapay zekaya eitim iin vereceimiz veriler:\n")
print(x.head())
```

```python
# train datamz ile test datamz ayryoruz
# train datamz sistemin salkl insan ile hasta insan ayrt etmesini renmek iin kullanlacak
# test datamz ise bakalm makine renme modelimiz doru bir ekilde hasta ve salkl insanlar ayrt edebiliyor mu diye 
# test etmek iin kullanlacak...
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1,random_state=1)

# knn modelimizi oluturuyoruz.
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print("K=3 iin Test verilerimizin dorulama testi sonucu ", knn.score(x_test, y_test))
```

```python

# k ka olmal ?
# en iyi k deerini belirleyelim..
sayac = 1
for k in range(1,11):
    knn_yeni = KNeighborsClassifier(n_neighbors = k)
    knn_yeni.fit(x_train,y_train)
    print(sayac, "  ", "Doruluk oran: %", knn_yeni.score(x_test,y_test)*100)
    sayac += 1
    
```
